```python
import SimpleITK as sitk
import nibabel as nib

img = sitk.ReadImage('/mnt/share/raheppt1/NAKO/AUTO_Tho_COR/070_sequence_2019-05-14T16-55-49_images.nii')
arr = sitk.GetArrayFromImage(img)


import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button
import numpy as np




def update(val):
    
    slice_index = int(slider.val)
    mri_slice.set_data(data[:,:,slice_index])
    label_slice.set_data(orig[:,:,slice_index])

    fig.canvas.draw_idle()
    
def toggle_label(event):
    label_slice.set_alpha(0.5)

    
def untoggle_label(event):
    label_slice.set_alpha(0)


def create_widgets():
    ax_slider = plt.axes([0.25, 0.03, 0.5, 0.03])
    ax_label = plt.axes([0.8, 0.05, 0.1, 0.03])
    ax_no_label = plt.axes([0.8, 0.02, 0.1, 0.03])

    slider = Slider(ax_slider, 'Slice', 0, data.shape[-1], valinit=0, valfmt="%i")

    button_toggle = Button(ax_label, "Label")
    button_untoggle = Button(ax_no_label, "No Label")
    
    return slider, button_toggle, button_untoggle

    
img = nib.load('/mnt/share/raheppt1/NAKO/AUTO_Tho_COR/070_sequence_2019-05-14T16-55-49_images.nii')
label = nib.load('/mnt/share/raheppt1/NAKO/AUTO_Tho_COR/070_sequence_2019-05-14T16-55-49_images.nii')
print(img.shape)
img = nib.load('/mnt/data/rawdata/NAKO_195/NAKO-195_MRT-Dateien/T1_3D_SAG_MPR_Tra_Defaced/111680_30_T1_3D_SAG_MPR_Tra (Defaced).nii.gz')
img = nib.load('/mnt/share/nora/imgdata/INBOX/11168030_undefined_111680_30/20160111093626_20160111/FLAIR_2D_TRA_ND26.nii')
plt.figure(figsize=(10,10))
data = img.get_fdata()
label = label.get_fdata()
orig = np.ma.masked_where(label == 0, label)

mri_slice = plt.imshow(data[:,:,0], cmap="gray")
label_slice = plt.imshow(orig[:, :, 0], alpha=0, cmap="coolwarm")


slider, button_toggle, button_untoggle = create_widgets()
slider.on_changed(update)
button_toggle.on_clicked(toggle_label)
button_untoggle.on_clicked(untoggle_label)
```

```python
import plotly.graph_objects as go
import numpy as np

# Create figure
fig = go.Figure()

# Add traces, one for each slider step
for step in np.arange(0, 5, 0.1):
    fig.add_trace(
        #go.Scatter(
        #    visible=False,
        #    line=dict(color="#00CED1", width=6),
        #    name="ðœˆ = " + str(step),
        #    x=np.arange(0, 10, 0.01),
        #    y=np.sin(step * np.arange(0, 10, 0.01))))
        go.Image(visible=False))

# Make 10th trace visible
fig.data[10].visible = True

# Create and add slider
steps = []
for i in range(len(fig.data)):
    step = dict(
        method="restyle",
        args=["visible", [False] * len(fig.data)],
    )
    step["args"][1][i] = True  # Toggle i'th trace to "visible"
    steps.append(step)

sliders = [dict(
    active=10,
    currentvalue={"prefix": "Frequency: "},
    pad={"t": 50},
    steps=steps
)]

fig.update_layout(
    sliders=sliders
)

fig.show()
```
