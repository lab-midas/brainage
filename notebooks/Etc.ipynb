```python
from pathlib import Path
import pydicom
import re
import shutil
```

```python
work_dir = Path('/mnt/data/rawdata/NAKO/NAKO_200/')
out_dir = Path('/mnt/share/raheppt1/NAKO/Dixon')

for directory in work_dir.glob('*3D_GRE_TRA_W_COMPOSE*'):
    print(directory)
    for file in directory.glob('**/*.*'):
        save_dir = out_dir.joinpath(directory.name)
        save_dir.mkdir(exist_ok=True)
        dataset = pydicom.read_file(str(file))
        series = str(dataset[0x0018,0x0022])
        echo_time = float(dataset.EchoTime)
        dataset.PatientName = str(dataset.PatientName).replace('_30', '')
        dataset.PatientID = str(dataset.PatientID).replace('_30', '')
        if 'DIXW' in series:
            save_dir = save_dir.joinpath('W')
            dataset.SeriesDescription = dataset.SeriesDescription + '_W'
            save_dir.mkdir(exist_ok=True)
            save_path = save_dir.joinpath(file.name)
        elif 'DIXF' in series:
            save_dir = save_dir.joinpath('F')
            save_dir.mkdir(exist_ok=True)
            dataset.SeriesDescription = dataset.SeriesDescription + '_F'
            save_path = save_dir.joinpath(file.name)
        elif 'SAT2' in series:
            if echo_time > 2:
                save_dir = save_dir.joinpath('INP')
                save_dir.mkdir(exist_ok=True)
                dataset.SeriesDescription = dataset.SeriesDescription + '_INP'
                save_path = save_dir.joinpath(file.name)
            else:
                save_dir = save_dir.joinpath('OPP')
                save_dir.mkdir(exist_ok=True)
                dataset.SeriesDescription = dataset.SeriesDescription + '_OPP'
                save_path = save_dir.joinpath(file.name)
        dataset.save_as(str(save_path))
```
