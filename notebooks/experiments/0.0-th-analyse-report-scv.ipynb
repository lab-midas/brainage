```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

## Simple regression network

#### IXI

```python
df_IXI_simple = pd.read_csv('/home/raheppt1/projects/age_prediction/reports/predictions/test_IXI_simple.csv')
df_IXI_simple
```

```python
df_select = df_IXI_simple[df_IXI_simple['name'].str.match('simple_IXI')]
df_subj_mean = df_select.groupby('subject').mean()
df_subj_std = df_select.groupby('subject').std()

# Mean prediction error
mae = mean_absolute_error(df_subj_mean['label'], df_subj_mean['prediction_0'])
rmse = math.sqrt(mean_squared_error(df_subj_mean['label'], df_subj_mean['prediction_0']))
print('EVALUATION METRICS')
print(f'mae {mae} rmse {rmse}')

plt.scatter(df_subj_mean['label'], df_subj_mean['prediction_0'], c='blue')
plt.title('predicted vs. chronological age')
plt.xlabel('chronological age')
plt.ylabel('predicted age (median)')
plt.ylim([10,90])
plt.show()
```

## Quantile regression

```python
df_IXI_quantile = pd.read_csv('/home/raheppt1/projects/age_prediction/reports/predictions/test_IXI_new_quantile_01.csv')
df_IXI_quantile
```

```python
df_select = df_IXI_quantile[df_IXI_quantile['name'].str.match('training_off_IXI')]
df_subj_mean = df_select.groupby('subject').mean()
df_subj_std = df_select.groupby('subject').std()

# Mean prediction error
mae = mean_absolute_error(df_subj_mean['label'], df_subj_mean['prediction_0'])
rmse = math.sqrt(mean_squared_error(df_subj_mean['label'], df_subj_mean['prediction_0']))
print('EVALUATION METRICS')
print(f'mae {mae} rmse {rmse}')
```

```python
plt.scatter(df_subj_mean['label'], df_subj_mean['prediction_0'], c='blue')
plt.title('predicted vs. chronological age')
plt.xlabel('chronological age')
plt.ylabel('predicted age (median)')
plt.ylim([10,90])
plt.show()

plt.scatter(df_subj_mean['label'], df_subj_mean['prediction_2'] - df_subj_mean['prediction_1'], c='orange')
plt.title('aleatoric uncertainty (quantile regression)')
plt.xlabel('chronological age')
plt.ylabel('predicted age (quantile difference)')
plt.xlim([10, 90])
plt.show()

plt.scatter(df_subj_mean['label'], df_subj_mean['prediction_2'] - df_subj_mean['prediction_0'], c='orange')
plt.scatter(df_subj_mean['label'], df_subj_mean['prediction_0'] - df_subj_mean['prediction_1'], c='red')
plt.title('aleatoric uncertainty (quantile regression)')
plt.xlabel('chronological age')
plt.ylabel('predicted age (quantile - median)')
plt.xlim([10, 90])
plt.show()

plt.scatter(df_subj_mean['label'], df_subj_mean['label'] - df_subj_mean['prediction_0'], c='red')
plt.title('prediction error vs. chronological age')
plt.xlabel('chronological age')
plt.ylabel('prediction error')
plt.xlim([10, 90])
plt.show()

plt.scatter( df_subj_mean['label'] - df_subj_mean['prediction_0'],  df_subj_mean['prediction_2'] - df_subj_mean['prediction_1'], c='red')
plt.title('uncertainty vs. prediction error')
plt.ylabel('predicted age (quantile difference)')
plt.xlabel('prediction error')
plt.show()

#plt.scatter(df_subj_mean['label'], df_subj_mean['prediction_2'])
#plt.errorbar(df_subj_mean['label'],df_subj_mean['prediction_0'],
#             yerr=2*df_subj_std['prediction_0'],
#             fmt='none')



#plt.scatter(df_subj_mean['label'], df_subj_mean['prediction_0'] - df_subj_mean['prediction_1'])
```

## Bayesian network

```python
df_IXI_bayesian = pd.read_csv('/home/raheppt1/projects/age_prediction/reports/predictions/test_IXI_bayesian_06.csv')
df_IXI_bayesian
```

```python
df_select = df_IXI_bayesian[df_IXI_bayesian['name'].str.match('training_off_ADNI_AD')] # training_on_random training_on_blurry1 training_off_IXI training_off_ADNI_NL training_off_ADNI_AD
df_subj_mean = df_select.groupby('subject').mean()
df_subj_std = df_select.groupby('subject').std()

# Mean prediction error
mae = mean_absolute_error(df_subj_mean['label'], df_subj_mean['prediction_0'])
rmse = math.sqrt(mean_squared_error(df_subj_mean['label'], df_subj_mean['prediction_0']))
mean_std = np.mean(df_subj_std['prediction_0'])

print('EVALUATION METRICS')
print(f'mae {mae} rmse {rmse}')
print(f'average std {mean_std}')

plt.scatter(df_subj_mean['label'], df_subj_mean['prediction_0'], c='blue')
plt.title('predicted vs. chronological age')
plt.xlabel('chronological age')
plt.ylabel('predicted age (mean)')
plt.xlim([10,90])
plt.show()

plt.scatter(df_subj_mean['label'], df_subj_mean['prediction_0'], c='blue')
plt.errorbar(df_subj_mean['label'],df_subj_mean['prediction_0'],
             yerr=2*df_subj_std['prediction_0'],
             fmt='none')
plt.title('predicted vs. chronological age')
plt.xlabel('chronological age')
plt.ylabel('predicted age (mean)')
#plt.xlim([10,90])
plt.show()
             
plt.scatter(df_subj_mean['label'], df_subj_std['prediction_0'], c='orange')
plt.title('predicted vs. chronological age')
plt.xlabel('chronological age')
plt.ylabel('predicted age (std)')
#plt.xlim([10,90])
plt.show()
```
